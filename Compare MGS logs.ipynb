```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import welly
from welly import Well,Project
from striplog import Legend, Striplog
```

### Stratigraphy ###

```python
#Import stratigraphy logs as strips:
legend = Legend.from_csv(filename='wells/legend.csv')              #import legend from csv
uwis = ['279454', '278586', '278815', '278816', '658966']          #list of unique well ids with geophysical and strat logs on Carleton campus, from west to east
strips = []                                                        #empty list to hold all strips once they are loaded
for uwi in uwis:                                                   #loop over unique well ids
    strip = Striplog.from_csv(filename='wells/'+uwi+'_strat_masl.csv')    #load striplog from stratigraphy csv file in meters
    strips.append(strip)
strips
```

```python
#Plot strat logs
f, ax = plt.subplots(1,len(uwis), figsize=(20,20), sharey=True)   #make a figure object with several axes for subplots, with the same y axis scale
for i,uwi in enumerate(uwis):                                     #loop over unique well ids and get counter
    strips[i].plot(ax=ax[i], legend=legend)                       #plot the current striplog
    ax[i].set_title(uwi)                                          #set the current title
    ax[i].set_ylim([123, 289])                                    #manually set the y axis limits so that all logs fit
```

## Caliper logs ##

```python
#Load data
uwis = ['279454', '278815', '278816', '658966']          #list of unique well ids with caliper logs on Carleton campus, from west to east
wells = []                                               #make empty list to store wells
for uwi in uwis:                                         #loop over unique well ids
    well = Project.from_las('wells/'+uwi+'_CALIPER*.las', index='m')[0]     # Load all las files in folder with specified well id in their name, and convert to meters
    wells.append(well)

#Convert from feet below land surface to meters above sea level
welldfs = []
for i,well in enumerate(wells):
    print(uwis[i])
    egl = well.header[well.header.mnemonic == 'EGL'].value.item() #get the value of the elevation above ground level from the header
    if isinstance(egl, str):
        print('oh no it is a string! look:', egl)
        egl = egl[:-2] #drop the last two characters
        print('fixed:',egl)
        egl = float(egl) #convert to float (drop last two characters first) 
    egl = egl*0.3048 #convert feet to meters
    
    caliper = well.data['CALIPER'].df  #pull out data for just the caliper log as a pandas dataframe
    caliper['ELEV'] = egl - caliper.index     #subtract depth (which is the index) from ground surface elevation (egl) to get elevation
    caliper.set_index('ELEV', inplace=True)   #replace the index (currently depth) with elevation
    welldfs.append(caliper)
```

```python
f, ax = plt.subplots(1,4, figsize=(20,15), sharey=True, sharex=True)   #make a figure object with several axes for subplots, with the same y axis scale
plt.suptitle('CALIPER')
ax[0].set_ylabel('Elevation [m a.s.l.]')

for i,uwi in enumerate(uwis):                                     #loop over unique well ids and get counter
    ax[i].plot(welldfs[i].CALIPER, welldfs[i].index)
    ax[i].set_title(uwi)                                          #set the current title
    #ax[i].set_ylim([123, 289])                                    #manually set the y axis limits so that all logs fit
```
